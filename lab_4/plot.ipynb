## Variant â„–10

$$\begin{cases}
-10.2x_1 + 9.3x_2 = 73 \\
5.3x_1 + 15.8x_2 = 173 \\
17.2x_1 + 10.3x_2 = 159 \\
\end{cases}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
equations = ['$-10.2x_1 + 9.3x_2 = 73$', '$5.3x_1 + 15.8x_2 = 173$', '$17.2x_1 + 10.3x_2 = 159$']

x_1 = np.linspace(-20, 20, 1000)

f = [(73 + 10.2 * x_1)/9.3, (173 - 5.3 * x_1)/15.8, (159 - 17.2 * x_1)/10.3]

fig, axes = plt.subplots(ncols=3, figsize=(10, 5))

for i, ax_ in enumerate(axes):
    ax_.set(xlim=(0, 20), ylim=(0, 20), title=equations[i])
    ax_.grid(linestyle=':', color='green')

axes[0].plot(x_1, f[0], '--',color='#34d2eb')
axes[1].plot(x_1, f[1], ':',color='#321652')
axes[2].plot(x_1, f[2], '-.',color='#385216')

folder = 'images/separate/'
plt.savefig(folder + 'equations.png')
plt.savefig(folder + 'equations.jpg')
plt.savefig(folder + 'equations.svg')
```

```python
matrices = [
    np.array([[-10.2, 9.3],[5.3, 15.8]]),
    np.array([[-10.2, 9.3],[17.2, 10.3]]),
    np.array([[5.3, 15.8],[17.2, 10.3]])
]
rhs = [
    np.array([73, 173]),
    np.array([73, 159]),
    np.array([173, 159])
]

solutions = [np.linalg.solve(matrices[i], rhs[i]) for i in range(len(matrices))]
```

```python
fig = plt.figure(figsize=(16, 9), dpi=100)
ax = plt.axes()
fig.add_axes(ax)
ax.set(xlim=(-2, 5), ylim=(5, 18), xlabel='$x_1$', ylabel='$x_2$', title='All lines in one figure')
ax.grid(linestyle='-.', color='green')

ax.plot(x_1, f[0], '-.c', label='f[0]')
ax.plot(x_1, f[1], '--y', label='f[1]')
ax.plot(x_1, f[2], ':m', label='f[2]')

atans = [math.degrees(math.atan(10.2 / 9.3)), math.degrees(math.atan(-5.3 / 15.8)), math.degrees(math.atan(-17.2 / 10.3))]
locations = [(-1, 7), (4, 9.8), (1, 14)]

ax.text(*locations[0], equations[0], rotation=atans[0], rotation_mode ='anchor', transform_rotates_text = True)
ax.text(*locations[1], equations[1], rotation=atans[1], rotation_mode ='anchor', transform_rotates_text = True)
ax.text(*locations[2], equations[2], rotation=atans[2], rotation_mode ='anchor', transform_rotates_text = True)

ax.plot(*solutions[0], marker='o', color='m')
ax.plot(*solutions[1], marker='o', color='r')
ax.plot(*solutions[2], marker='o', color='c')

plt.fill_between(x_1, np.minimum(f[0], f[2]), np.minimum(np.minimum(f[0], f[2]), f[1]), color='grey', alpha=0.5)

ax.annotate(f"The intersection point of\n f[0] and f[1]", ha="center", xy=solutions[0], xytext=(0.5, 9), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[1] and f[2]", ha="center", xy=solutions[2], xytext=(3, 8), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[0] and f[2]", ha="center", xy=solutions[1], xytext=(3, 14), arrowprops=dict(arrowstyle="->"))

ax.legend()

folder = 'images/together/'
plt.savefig(folder + 'equations.png')
plt.savefig(folder + 'equations.jpg')
plt.savefig(folder + 'equations.svg')

plt.show()
```
